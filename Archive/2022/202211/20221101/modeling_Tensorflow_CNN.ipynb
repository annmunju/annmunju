```python
import tensorflow as tf
tfk = tf.keras # use datasets, layers, models
```

```python
(train_images, train_labels), (test_images, test_labels) = tfk.datasets.mnist.load_data()
```

```python
train_images = train_images.reshape((60000, 28, 28, 1))
test_images = test_images.reshape((10000, 28, 28, 1))

# 픽셀 값을 0~1 사이로 정규화합니다.
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python
model = tfk.models.Sequential()
model.add(tfk.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(tfk.layers.MaxPooling2D((2, 2)))
model.add(tfk.layers.Conv2D(64, (3, 3), activation='relu'))
model.add(tfk.layers.MaxPooling2D((2, 2)))
model.add(tfk.layers.Conv2D(64, (3, 3), activation='relu'))
model.add(tfk.layers.Flatten())

# 3D -> 1D로 펼침
model.add(tfk.layers.Dense(64, activation='relu'))

# 10개의 output값 (0~9까지의 분류)
model.add(tfk.layers.Dense(10, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(train_images, train_labels, epochs=5)
```

```python
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
```
