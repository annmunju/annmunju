```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
entropy = lambda x: -x*(np.log2(x)) -(1-x)*(np.log2(1-x))
```

```python
entropy(0)
```

```python
entropy(1)
```

```python
x = np.linspace(0.01, 0.99, 100)
```

```python
y = entropy(x)
```

```python
plt.plot(x,y,'o')
```
