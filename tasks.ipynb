```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10, 10, 20)
x
```

```python
y = np.log(x)
y
```

```python
plt.scatter(x, y)
plt.xlim((-10, 10))
plt.ylim((-10, 10))
plt.show()
```

```python
y_2 = -np.log(x)
y_2
```

```python
plt.scatter(x, y_2)
plt.xlim((-10, 10))
plt.ylim((-10, 10))
plt.show()
```

```python
y_3 = -np.log((1-x))
y_3
```

```python
plt.scatter(x, y_3)
plt.xlim((-10, 10))
plt.ylim((-10, 10))
plt.show()
```

```python
plt.plot(x, y)
plt.plot(x, y_2)
plt.plot(x, y_3)
plt.legend(('log', '-log', '-log(-x)'))

plt.show()
```
